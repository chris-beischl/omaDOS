```python
from omados.engine.cards import Cards
from omados.engine.rules import can_play_sauspiel, get_available_games

# Test hand with known properties
hand = Cards(["EO", "GO", "HO", "SO", "EU", "GU", "HU", "SU"])  # all obers and unters
print("All Obers+Unters:")
print(f"  can_play_sauspiel: {can_play_sauspiel(hand)}")
print(f"  available_games: {get_available_games(hand, caller_id=0)}")

# Hand with Eichel-Sau and some Eichel cards -> cannot call Eichel
hand2 = Cards(["EA", "EK", "E9", "SA", "SK", "S9", "GA", "GK"])
print(
    "\nHand with EA (cannot call Eichel), SA (cannot call Schellen),"
    " GA (cannot call Gras):"
)
print(f"  can_play_sauspiel: {can_play_sauspiel(hand2)}")
print(f"  available_games count: {len(get_available_games(hand2, caller_id=0))}")

# Hand that can call Eichel
hand3 = Cards(["EK", "E9", "SA", "SK", "GA", "GK", "HO", "HU"])
print("\nHand with EK but no EA -> can call Eichel:")
print(f"  can_play_sauspiel: {can_play_sauspiel(hand3)}")
```
