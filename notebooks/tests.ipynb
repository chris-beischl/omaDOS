#### Game sequence: 
1. deal cards 
2. every player decides for themselves if they want to play and what
3. circularly ask if they want to play 
    1. yes: ask others if they want to play solo 
        1. yes: move to solo decision process
        2. no: let player decide color they want to play
    2. no: move on to next player
    3. default: play Ramsch
4. start playing
    1. first player or player who got last stich starts
    2. move clockwise, each player gets to play a card
    3. determine winner of stich
    4. update states
5. team wins which got over 60 points
