```python
%load_ext autoreload
%autoreload 2

# import logging

from omados.agents.greedy import BaseAgent, GreedyAgent
from omados.agents.random import RandomAgent
from omados.engine.cards import Cards
from omados.env.game import SchafkopfEnv

# logging.basicConfig(level=logging.DEBUG)
```

```python
agents: list[BaseAgent] = [GreedyAgent(player_id) for player_id in range(4)]
env = SchafkopfEnv(agents=agents)

errors = 0
for _ in range(1000):
    try:
        env.play_game()
    except Exception as e:
        errors += 1
        print(f"Error during game: {e}")
print(f"Completed 1000 games with {errors} errors.")
```
